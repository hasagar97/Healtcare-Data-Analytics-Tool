```python
def add_column(df,col_info):
    print(col_info)
    if col_info["dataType"].lower() == "categorical":
        df[col_info["Header"]] = np.random.choice(col_info["Values"], df.shape[0])
    elif col_info["dataType"].lower() == "numerical":
        df[col_info["Header"]] =np.random.randint(col_info["Values"][0], col_info["Values"][1], df.shape[0])
    return df
```

```python
def get_dummy_df(n,metaD):
    df = pd.DataFrame(data = np.transpose([np.arange(1,n+1)]), columns = ["num"])
    df2 = df
    for d in metaD:
        df2 = add_column(df2,d)
    df2["target_from"] = df["target_to_from"].str.split('_').str[0]
    df2["target_to"] = df["target_to_from"].str.split('_').str[1]
    return df2
```
